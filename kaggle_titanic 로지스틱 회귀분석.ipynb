```python
import pandas as pd
import numpy as np
import matplotlib as plt
%matplotlib inline
```

```python
train_df = pd.read_csv("D:/py_fi/titanic/train.csv")
test_df = pd.read_csv("D:/py_fi/titanic/test.csv")
```

```python
train_df.head()
```

```python
train_df.set_index('PassengerId', inplace = True)
test_df.set_index("PassengerId",inplace = True)
```

```python
train_df.head()
```

```python
train_index = train_df.index
test_index = test_df.index
```

```python
y_train_df = train_df.pop("Survived")
```

```python
y_train_df.head()
```

## Data Preproecessing

```python
#플로팅을 소수점 두자리까지만 보여줌
pd.set_option('display.float_format', lambda x : '%.2f' %x)
```

```python
test_df.isnull().sum() / len(test_df)
```

```python
test_df.isnull().sum() / len(test_df) * 100
```

```python
del train_df["Cabin"]
del test_df["Cabin"]
```

```python
all_df = train_df.append(test_df)
all_df
```

```python
all_df.isnull().sum() / len(all_df) * 100
```

```python
del all_df["Name"]
del all_df["Ticket"]
```

```python
all_df.head()
```

```python
all_df["Sex"] = all_df["Sex"].replace({"male":0,"female":1})
all_df.head()
```

```python
all_df["Embarked"].unique()
```

```python
all_df["Embarked"] = all_df["Embarked"].replace({"S":0,"C":1,"Q":2,np.nan:99})
```

```python
all_df["Embarked"].unique()
```

```python
pd.get_dummies(all_df["Embarked"],prefix = "embarked")
```

```python
matrix_df = pd.merge(all_df, pd.get_dummies(all_df["Embarked"],prefix = "embarked"), left_index = True, right_index=True)
matrix_df.head()
```

```python
matrix_df.corr()
#상관관계 분석
```

```python
#그룹함수로 클래스별의 나이 평균을 계산
all_df.groupby("Pclass")["Age"].mean()
```

```python
#Pclass에 탑승하고 나이가 null값은 사람들을 출력
all_df.loc[ (all_df["Pclass"] == 1) & (all_df["Age"].isnull()) , "Age"]
```

```python
#결측치를 평균값으로 대체
all_df.loc[(all_df["Pclass"] == 1) & (
        all_df["Age"].isnull()), "Age"] = 39.16
all_df.loc[ (
    all_df["Pclass"] == 2) & all_df["Age"].isnull() , "Age"] = 29.51
all_df.loc[ ( all_df["Pclass"] == 3) & all_df["Age"].isnull(), "Age"] = 24.82
```

```python
all_df.isnull().sum()
```

```python
#클래스별 Fare의 평균계산 
all_df.groupby("Pclass")["Fare"].mean()
```

```python
all_df[all_df["Fare"].isnull()]
```

```python
all_df.loc[all_df["Fare"].isnull(), "Fare"] = 13.30
```

```python
del all_df["Embarked"]
```

```python
all_df.head()
```

```python
all_df["Pclass"] = all_df["Pclass"].replace({1:"A",2:"B",3:"C"})
```

```python
all_df.head()
```

```python
all_df = pd.get_dummies(all_df)
all_df.head()
```

```python
all_df = pd.merge(all_df,
                  matrix_df[["embarked_0","embarked_1","embarked_2","embarked_99"]],
                 left_index = True, right_index = True)
```

```python
all_df.head()
```

```python
train_df = all_df[all_df.index.isin(train_index)]
test_df = all_df[all_df.index.isin(test_index)]
```

```python
train_df.head()
test_df.head()
```

# Build Model

```python
x_data = train_df.as_matrix()
y_data = y_train_df.as_matrix()
```

```python
x_data.shape, y_data.shape
```

```python
y_data
```

```python
from sklearn.linear_model import LogisticRegression 
cls = LogisticRegression()
cls.fit(x_data,y_data)
```

```python
cls.intercept_
```

```python
cls.coef_
```

```python
cls.predict(test_df.values)
```

```python
test_df.index
```

```python
x_test = test_df.as_matrix()
y_test = cls.predict(x_test)
y_test
```

```python
result = np.concatenate((test_index.values.reshape(-1,1),
                        cls.predict(x_test).reshape(-1,1)),axis = 1)
result[:5]
```

```python
df_submission = pd.DataFrame(result, columns=["PassengerId","Survived"])
df_submission
```

```python
df_submission.to_csv("submission_result.csv",index=False)
```
